```python
# If you haven't installed yfinance yet, uncomment the next line:
# !pip install yfinance

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Define the ticker symbol and time frame
ticker_symbol = 'AAPL'
start_date = '2020-01-01'
end_date = '2020-12-31'

# Download historical data
data = yf.download(ticker_symbol, start=start_date, end=end_date,auto_adjust=False)

# Display the first few rows of the downloaded data
print("Downloaded Data:")
print(data.head())

# The downloaded data contains these default columns:
# ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']

# Calculate derived metrics:

# 1. Daily Returns
data['Daily Return'] = data['Adj Close'].pct_change()

# 2. Moving Averages (e.g., 20-day moving average)
data['MA20'] = data['Adj Close'].rolling(window=20).mean()

# 3. Volatility: standard deviation of daily returns (rolling 20-day volatility)
data['Volatility'] = data['Daily Return'].rolling(window=20).std()

print("\nData with Derived Metrics:")
print(data[['Adj Close', 'Daily Return', 'MA20', 'Volatility']].head(25))

# (Optional) Plotting the Adjusted Close and Moving Average
plt.figure(figsize=(10, 5))
plt.plot(data.index, data['Adj Close'], label='Adj Close')
plt.plot(data.index, data['MA20'], label='20-Day MA', linestyle='--')
plt.title(f'{ticker_symbol} Adjusted Close and 20-Day Moving Average')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

# (Optional) Downloading additional data like dividends or splits:
ticker = yf.Ticker(ticker_symbol)
dividends = ticker.dividends
splits = ticker.splits

print("\nDividends:")
print(dividends.head())

print("\nStock Splits:")
print(splits.head())

# Note:
# Market capitalization is not directly available in the historical data download.
# It may be accessed through the Ticker.info attribute as a snapshot:
info = ticker.info
market_cap = info.get('marketCap', 'N/A')
print(f"\nMarket Capitalization (Current): {market_cap}")
```

## Actual Dataset gathering starts from here

```python
import yfinance as yf
import pandas as pd
```

```python
# List of target tickers
tickers = ['AAPL', 'TSLA', 'AMZN', 'NKE', 'NVDA']
```

```python
# Define your time frame (modify as needed)
start_date = '2025-01-01'
end_date = '2025-03-31'
```

```python
def download_stock_data(ticker, start, end):
    """
    Downloads historical stock data and computes derived metrics.
    """
    stock = yf.Ticker(ticker)
    # Download historical data (Open, High, Low, Close, Adj Close, Volume)
    df = stock.history(start=start, end=end, interval="1d", auto_adjust=False)
    
    if df.empty:
        print(f"No data found for {ticker}")
        return df

    # Compute derived metrics:
    # Daily returns based on Adjusted Close
    df['Daily Return'] = df['Adj Close'].pct_change()
    
    # Example moving averages (20-day and 50-day)
    df['MA20'] = df['Adj Close'].rolling(window=20).mean()
    df['MA50'] = df['Adj Close'].rolling(window=50).mean()
    
    # Volatility: rolling standard deviation of daily returns (20-day window)
    df['Volatility'] = df['Daily Return'].rolling(window=20).std()
    
    # Additional info: market capitalization and dividend yield (static info)
    info = stock.info
    df['Market Cap'] = info.get('marketCap')
    df['Dividend Yield'] = info.get('dividendYield')
    
    # Reset index to make Date a column
    df = df.reset_index()
    
    return df
```

```python
# Dictionary to hold dataframes for each ticker
data_dict = {}

for ticker in tickers:
    print(f"Downloading data for {ticker}...")
    df = download_stock_data(ticker, start_date, end_date)
    data_dict[ticker] = df
    # Optionally, save to CSV
    df.to_csv(f"../results0/{ticker}_yahoo_data_0.csv", index=False)

print("Stock data downloaded and saved successfully.")
```

## Trying to Downloading News articles related to above stocks on each day.

```python
import yfinance as yf
import pandas as pd
import requests
from datetime import datetime
import time
```

```python
# Define your target tickers and date range for news matching the stock data dates.
tickers = ['AAPL', 'TSLA', 'AMZN', 'NKE', 'NVDA']
```

```python
# Set the date range to match your historical stock data
start_date = pd.to_datetime('2023-01-01')
end_date = pd.to_datetime('2023-03-31')
```

```python
# Function to convert Unix timestamp to date
def timestamp_to_date(ts):
    return datetime.fromtimestamp(ts).date()
```

```python
def get_news_for_ticker(ticker, start, end):
    """
    Retrieves news for a given ticker and organizes them by date.
    Returns a dictionary with dates as keys and a list of news articles as values.
    Each news article contains title, summary (if available), and publication date.
    """
    stock = yf.Ticker(ticker)
    news_list = stock.get_news(count=10000)
    print(len(news_list))
    
    # Initialize a dictionary to store news articles by date
    news_by_date = {}
    
    # Iterate over news items
    for news in news_list:
        # Some news items may not have a publication timestamp
        pub_ts = news.get('providerPublishTime')
        if not pub_ts:
            continue
        
        pub_date = timestamp_to_date(pub_ts)
        
        # Filter news within the desired date range
        if start.date() <= pub_date <= end.date():
            # Extract required fields; note that "summary" may be under 'summary' or 'description'
            title = news.get('title', '')
            # Some items may have a 'summary' key; if not, you can check for 'description'
            summary = news.get('summary', news.get('description', ''))
            
            # Append news article to the date key
            if pub_date not in news_by_date:
                news_by_date[pub_date] = []
            if len(news_by_date[pub_date]) < 10:  # limit to 10 articles per day
                news_by_date[pub_date].append({
                    'title': title,
                    'summary': summary,
                    'publication_date': pub_date
                })
    
    return news_by_date
```

```python
# Dictionary to hold news data for each ticker
news_data = {}

for ticker in tickers:
    print(f"Retrieving news for {ticker}...")
    news_by_date = get_news_for_ticker(ticker, start_date, end_date)
    news_data[ticker] = news_by_date
    # Optionally, save the news data to a CSV per ticker
    rows = []
    for news_date, articles in news_by_date.items():
        for article in articles:
            rows.append({
                'ticker': ticker,
                'date': news_date,
                'title': article['title'],
                'summary': article['summary']
            })
    news_df = pd.DataFrame(rows)
    time.sleep(5)
    news_df.to_csv(f"{ticker}_news_data.csv", index=False)

print("News data downloaded and saved successfully.")
```
